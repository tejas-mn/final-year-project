# Transform images into different types ( b&w , segmented) to use in the modeling phase later on

**Segmentation**

```python
def segment_image(file):
    img = cv2.imread(file)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    hsv_img = cv2.cvtColor(img, cv2.COLOR_RGB2HSV)
    lower_green = np.array([25,0,20])
    upper_green = np.array([100,255,255])
    mask = cv2.inRange(hsv_img, lower_green, upper_green)
    result = cv2.bitwise_and(img, img, mask=mask)
    lower_brown = np.array([10,0,10])
    upper_brown = np.array([30,255,255])
    disease_mask = cv2.inRange(hsv_img, lower_brown, upper_brown)
    disease_result = cv2.bitwise_and(img, img, mask=disease_mask)
    final_mask = mask + disease_mask
    final_result = cv2.bitwise_and(img, img, mask=final_mask)
    output_path='/Segmented/'+os.path.basename(os.path.dirname(file))+'/'+os.path.basename(file)
    cv2.imwrite("output_pat.jpg",final_result)
```

```python
segment_image("images.jpg")
```

```python
def convert_dataset(rootdir,type):
  '''
  function to convert every file in subfolders into segmented images ready to be used in a script
  '''
  count=0
  for subdir, dirs, files in os.walk(rootdir):
    for file in files:
      path=str(subdir)+'/'+str(file)
      if type == 'Color':
        segment_image(path)
        count+=1
      else:
        segment_image_HSV(path)
        count+=1

  print('segmented '+str(count)+' images correctly')  
```

```python
convert_dataset('/content/drive/MyDrive/datasetCBL/Color','Color')
```

#HSV segmentation


```python
def segment_image_HSV(file):
  '''
  function to convert image into a segmented image in the HSV color space ready to use in a script
  '''
  img = cv2.imread(file)
  hsv_img = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
  output_path='/content/drive/MyDrive/datasetCBL/Segmented_HSV/'+os.path.basename(os.path.dirname(file))+'/'+os.path.basename(file)
  cv2.imwrite(output_path,hsv_img)
```

```python
convert_dataset('/content/drive/MyDrive/datasetCBL/Segmented','HSV')
```

```python
#reading the dataset from the directory
Dataset_HSV = tf.keras.preprocessing.image_dataset_from_directory(
    "/content/drive/MyDrive/datasetCBL/Segmented_HSV",
    shuffle=True,
    image_size = (IMAGE_SIZE,IMAGE_SIZE),
    batch_size = BATCH_SIZE
)
```
